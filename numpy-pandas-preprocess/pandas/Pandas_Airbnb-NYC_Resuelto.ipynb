**6.	Crea una nueva columna con el rating de acuerdo a la siguiente formula, donde las variables son: puntuación del municipio, número de reviews y disponibilidad de la habitación al año.**

**Formula: Rating = (Puntuación de ciudad) + (número de review/ 629) + (disponibilidad / 365).**

La puntuación para cada ciudad es la siguiente {"Manhattan":10/10,"Brooklyn":8/10,"Queens":6/10,"Staten Island":4/10,"Bronx":2/10}.

```python
puntuaje_ciudad={"Manhattan":10/10,"Brooklyn":8/10,"Queens":6/10,"Staten Island":4/10,"Bronx":2/10}
puntuaje_ciudad
```

```python
list_rating=[]
for i in df.index:
    ciudad = df.loc[i]["municipio"]
    num_reviews = df.loc[i]["number_of_reviews"]
    disponibilidad = df.loc[i]["availability_365"]
    rating =  puntuaje_ciudad[ciudad]+ num_reviews/629 + disponibilidad/365
    list_rating.append(rating)
#list_rating

```

```python
df["rating"] = list_rating
df.head()
```

```python
puntajes_ciudad = {"Manhattan":10/10,"Brooklyn":8/10,"Queens":6/10,"Staten Island":4/10,"Bronx":2/10}
df['Rating2'] = df.apply(lambda x: puntajes_ciudad[x['municipio']] + x['number_of_reviews']/629 + x['availability_365']/365,axis=1)
df.head()
```

```python
def rating(municipio,number_of_reviews,availability_365):
    if "Manhattan" in municipio:
        return (10/10)+(number_of_reviews/629)+(availability_365/365)
    elif "Brooklyn" in municipio:
        return (8/10)+(number_of_reviews/629)+(availability_365/365)
    elif "Queens" in municipio:
        return (6/10)+(number_of_reviews/629)+(availability_365/365)
    elif "Staten Island" in municipio:
        return (4/10)+(number_of_reviews/629)+(availability_365/365)
    else:
        return (2/10)+(number_of_reviews/629)+(availability_365/365)
    
df["Rating56"]=df.apply(lambda x: rating(x["municipio"],x["number_of_reviews"],x["availability_365"]),axis=1)
df
```

```python
df[["municipio","number_of_reviews","availability_365"]].apply(rating)
```
