**18. Agrega una nueva columna al daframe que se llame: "Rango edad" y va a catagorizar de la siguiente manera:**
- Niño < 11 años
- Joven <18 años
- Adulto < 50 años
- Mayor >= 50 años

```python
def fun(x):
    if x < 11:
        return "Niño"
    elif 11 <= x < 17:
        return "Joven"
    elif 18<= x < 49:
        return "Adulto"
    else:
        return "Mayor"
```

```python
df["Rango Edad"] = df["Age"].map(fun)
df
```

```python
df.loc[df['Age'] < 11, 'Rango edad'] ='Niño'
df.loc[df['Age'].between(11,17), 'Rango edad'] = 'Joven'
df.loc[df['Age'].between(18, 49), 'Rango edad'] = 'Adulto'
df.loc[df['Age']>=50, 'Rango edad']='Mayor'
df
```
