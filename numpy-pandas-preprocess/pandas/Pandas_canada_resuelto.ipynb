###  Obtener un dataframe que te parezca relevante a partir de la aplicaciÃ³n de filtros. Â¡Se creativo!

```python
media_paises=(df.iloc[:,4:-1].mean(axis=1)).tolist()
std=(df.iloc[:,4:-1].std(axis=1)).tolist()
maximo=(df.iloc[:,4:-1].max(axis=1)).tolist()
minimo=(df.iloc[:,4:-1].min(axis=1)).tolist()
df3 =pd.DataFrame(columns = ["Continente","Pais","Total Migracion","Media","Maximo","Minimo","std"])
df3["Pais"]=df["Pais"]
df3["Continente"]=df["Continente"]
df3["Total Migracion"]=df["Total migracion"]
df3["Media"]=media_paises
df3["std"]=std
df3["Maximo"]=maximo
df3["Minimo"]=minimo
df3=df3.sort_values("Continente")
df3.reset_index(drop=True)
africa=df3[df3["Continente"]=="Africa"]
africa=africa.sort_values("Total Migracion",ascending=False)

sudamerica=df3[df3["Continente"]=="Sudamerica"]
sudamerica=sudamerica.sort_values("Total Migracion",ascending=False)

norteamerica=df3[df3["Continente"]=="Norteamerica"]
norteamerica=norteamerica.sort_values("Total Migracion",ascending=False)


europa=df3[df3["Continente"]=="Europa"]
europa=europa.sort_values("Total Migracion",ascending=False)

asia=df3[df3["Continente"]=="Asia"]
asia=asia.sort_values("Total Migracion",ascending=False)

oceania=df3[df3["Continente"]=="Oceania"]
oceania=oceania.sort_values("Total Migracion",ascending=False)

#pd.concat(africa,sudamerica,norteamerica,europa,asia,oceania)
oceania
```
